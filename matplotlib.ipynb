```python
np.random.seed(0)
arr = np.random.standard_normal((8, 100))

plt.subplot(2, 2, 1)
plt.scatter(arr[0], arr[1], c = arr[1])
plt.spring()
plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c = arr[3])
plt.summer()
plt.title('summer')

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c = arr[5])
plt.autumn()
plt.title('autumn')

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c = arr[7])
plt.winter()
plt.title('winter')

plt.tight_layout()
```
